```python
class learning_uploaded_csv(user_name):
    def __init__(self,name):
        self.user_name = user_name
        import pandas as pd
        import tensorflow as tf
        from sklearn.cross_validation import train_test_split
        from sklearn.preprocessing import MinMaxScaler
        from sklearn.metrics import mean_squared_error
    
    #必要なモジュールのインポート
    def read_csv(csv,features,target_value):
        #pd.read_csvでアップロードされたcsvファイルを読み込む
        df = pd.read_csv(csv)
        
        #ボックスに入力した値だけのデータフレームを作成する。
        #X = df.drop('input()',axis=1)
        #y_true = df['input()']
        X = df.drop(target_value,axis=1)
        y_true = df[features]
    
    #与えられたCSVからtrainingさせるものとテストに使用するものを分割する。パラメータはユーザが決められる。
    #X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=(input()))
    def train_test_split(test_size):
        X_train, X_test, y_train, y_test = train_test_split(X, y_true, test_size=test_size)
    
    #正規化させる
    def MinMaxScaler(X_train,X_test):
        
        scaler = MinMaxScaler()
        scaler.fit(X_train)
        
        X_train = pd.DataFrame(scaler.transform(X_train),columns=X_train.columns,index=X_train.index)
        X_test= pd.DataFrame(scaler.transform(X_test),columns=X_test.columns,index=X_test.index)
    
    #与えられたCSVの列名をリストに入力する。
    def creating_numeric_feature_columns(X):
        feat_col = [] 
        X_columns = X.columns
    #列名をもとにfeature_columnを作成する。変数名は列名そのまま
        for i in X_columns:
            feat_col.append(tf.feature_column.numeric_column('{}'.format(i)))
    
    #学習させるためのinputを作成する
    #input_func = tf.estimator.inputs.pandas_input_fn(X_train,y_train,num_epochs=input(),shuffle=input())
    def create_input_func(X_train,y_train,num_epochs):
        input_func = tf.estimator.inputs.pandas_input_fn(X_train=X_train,y_train=y_train,num_epochs=num_epochs,shuffle=True)
    
    #学習モデルの作成。プルダウンなどから学習モデルの種別やパラメータを選択する
    #model = tf.estimator.(DNNRegressor)(input([10,10,10,10,10]),feat_col)
    def create_DNNRegressor_model():
        model = tf.estimator.DNNRegressor([5,5,5],feat_col)
    
    #パラメータを選択し学習開始ボタンをクリックすると学習が開始される。
    #model.train(input_func,input(steps=300000))
    def model_train(input_func,steps):
        model.train(input_func,steps=steps)
    
    #以降はユーザが作成した学習モデルをもとに予測する。
    #予測するためのfunctionを作成する。ユーザはパラメータを決定できる
    #pred_func = tf.estimator.inputs.pandas_input_fn(X_test,y_test,shuffle=False,batch_size=input(1),num_epochs=input(1))
    def pred_func():
        final_pred = []
        pred_func = tf.estimator.inputs.pandas_input_fn(X_test,y_test,shuffle=False,batch_size=1,num_epochs=1)
        #見やすいように予測をリスト変換
        prediction = list(model.predict(pred_func))
        
        for i in prediction:
            final_pred.append(i['predictions'])
        return mean_squared_error(y_test,final_pred) ** 0.5
    
    
    #ユーザが画面から入力したデータを処理し、予測を出力する
    def user_inputdata_predictions(user_input_data):
        user_data = pd.DataFrame(columns=X.columns)
        for i in range(1):
            user_data.loc[i] = [ x for x in user_input_data]
        test_func = tf.estimator.inputs.pandas_input_fn(user_data,shuffle=False,batch_size=1,num_epochs=1)
        user_prediction = list(model.predict(test_func))
        return user_prediction[0]['predictions']
```
